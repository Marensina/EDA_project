```python

df_combined = pd.concat([yarrowpoint.assign(dataset='Yarrow Point'), educationhill.assign(dataset='Education Hill'), 
                         Factoria.assign(dataset='Factoria'), redmond.assign(dataset='Redmond'), 
                         sammamish.assign(dataset='Sammamish'), west_bellevue.assign(dataset='West Bellevue'), 
                         Unionhill.assign(dataset='Union Hill')])

# Define markers for each dataset
markers = {'Yarrow Point': 'o', 'Education Hill': '*', 'Factoria': 'D', 'Redmond': '^', 
           'Sammamish': 'v', 'West Bellevue': 'P', 'Union Hill': 'X'}
colors = ['red', 'green', 'blue', 'purple', 'orange', 'brown', 'pink']

# Plot scatter plot using Seaborn with different markers for each dataset
sns.scatterplot(data=df_combined, x='price', y='bedrooms', hue='dataset', style='dataset', markers=markers, palette=colors)

# Add labels and title
plt.xlabel('Price',fontsize=12, fontstyle='italic',fontname='Times New Roman')
plt.ylabel('Bedrooms',fontsize=12, fontstyle='italic',fontname='Times New Roman')
plt.title('Scatter Plot for Multiple DataFrames',fontsize=14, fontstyle='italic',fontname='Times New Roman')
plt.legend(title='Areas:')
# Show plot
plt.show()
```

```python
# Define markers for each dataset
markers = {'Yarrow Point': 'o', 'Education Hill': '*', 'Factoria': 'D', 'Redmond': '^', 
           'Sammamish': 'v', 'West Bellevue': 'P', 'Union Hill': 'X'}
colors = ['red', 'green', 'blue', 'purple', 'orange', 'brown', 'red']

# Create subplots
fig, axes = plt.subplots(3, 3, figsize=(15, 15))  # 3x3 grid of subplots

# Flatten axes for easier indexing
axes = axes.flatten()

# Plot each dataset on a separate subplot
for ax, (name, group), color in zip(axes, df_combined.groupby('dataset'), colors):
    sns.scatterplot(data=group, x='price', y='bedrooms', ax=ax, color=color, marker=markers[name])
    ax.set_title(name)  # Set subplot title
    ax.set_xlabel('Price')  # Set xlabel
    ax.set_ylabel('Bedrooms')  # Set ylabel
    
    # Adjust y-axis ticks size and format
    ax.tick_params(axis='y', labelsize=10)  # Set y-axis tick label size
# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
